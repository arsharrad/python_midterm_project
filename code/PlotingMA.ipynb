```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
INPUT_DIR = os.path.join('../data')
```

```python
XOM_PATH = os.path.join(INPUT_DIR, 'XOM.csv')
```

```python
def import_cvs_as_dataframe(PATH, df):
    ''' Takes the relative path to dataset and import it as a clean data frame '''
    df = pd.read_csv(PATH)
    df = df.dropna()
```

```python
def plot_fast_slow_signals(df):
    ''' This function takes a data frame of a certain stock, mutate it by calculating the
    fast signal (Moving average 7 days) and the slow signal (Moving Average 40 days), 
    and return a plot to visualize the signals of the stock.
    '''
    
    df['MA7'] = df['Close'].rolling(7).mean()
    df['MA40'] = df['Close'].rolling(40).mean()
    
    plt.figure(figsize=(10, 8))
    df['MA7'].loc['2021-01-01':'2021-12-31'].plot(label='Fast Signal')
    df['MA40'].loc['2021-01-01':'2021-12-31'].plot(label='Slow Signal')
    df['Close'].loc['2021-01-01':'2021-12-31'].plot(label='Closing Price')
    plt.legend()

    return plt.show()
```
